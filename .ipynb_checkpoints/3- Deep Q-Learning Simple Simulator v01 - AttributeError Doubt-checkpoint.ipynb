```python
import numpy.random as rnd
import numpy as np
import tensorflow as tf
import tensorflow.contrib.slim as slim
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from tensorflow import keras
from IPython.display import Image
Image(filename='fig01.PNG')
```

```python
discount_rate = 0.95 # gamma
learning_rate0 = 0.05
learning_rate_decay = 0.1
n_iterations = 20000

# numero de estados
s_num= 4
# numero de acoes
a_num= 4

print("Numero de Estados: ", s_num)
print("Numero de Acoes: ", a_num) 

#Q = np.array(np.zeros([s_num,a_num]))

model = keras.Sequential()
model.add(keras.layers.Dense(s_num, input_shape=(s_num,), activation='relu'))
model.add(keras.layers.Dense(a_num, activation='softmax'))
model.compile(optimizer=tf.train.AdamOptimizer(0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
def choose_action(s, epsilon, numChannels=4):
# Selecione a ao com o maior valor Q.
    if np.random.uniform() < epsilon:
        a = np.argmax(model.predict(s)[0])
# Selecione uma ao aleatriamente.
    else:
        a = np.random.randint(0,numChannels)
    return a
```
