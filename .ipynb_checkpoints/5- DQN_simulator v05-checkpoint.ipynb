# Recomendação da apresentação da DeepMind

- Utilizar: *RMSPropOptimizer*, *Huber Loss* e ajustar a *taxa de exploração* (começando de 1 e decremetar por 0.1 ou 0.05).
-  RMSPropOptimizer Arguments:
        learning_rate: `float`. Learning rate.
        decay: `float`. Discounting factor for the history/coming gradient.
        momentum: `float`. Momentum.
        epsilon: `float`. Small value to avoid zero denominator.
        use_locking: `bool`. If True use locks for update operation.
        name: `str`. Optional name prefix for the operations created when
            applying gradients. Defaults to "RMSProp".
            
- Huber Loss: é menos sensível a outliers(lies outside, pontos fora da curva, dados atípicos/discrepantes) em comparação com o MSE e MAE. No Huber loss o argumento Delta deve ser especificado e ajustado. Quando o Erro é menor que o Delta é um erro quadrático (MSE), caso o erro seja maior que Delta é um erro absoluto (MAE)
        
- Anteriormente estava a utilizar o Adam optimizer e categorical_crossentropy loss

```python
# Definição do Modelo da Rede Neural
def create_neural_network_model(state_size, action_size):
    model = keras.Sequential()
    model.add(keras.layers.Dense(state_size, input_shape=(state_size,), activation='relu'))
    model.add(keras.layers.Dense(action_size, activation='softmax'))
    #model.compile(optimizer=tf.train.AdamOptimizer(0.001),loss='categorical_crossentropy',metrics=['accuracy'])
    model.compile(optimizer=tf.train.RMSPropOptimizer(0.00025, 0.99, 0.0, 1e-6), loss= keras.losses.Huber(delta=1.0, name='huber_loss'), metrics=['accuracy']) 
    return model

def choose_action(c_state, epsilon, state_size):
   # Selecione a ação com o maior valor Q.
    if np.random.uniform() < epsilon:
        predicted = prediction_network.predict(c_state.reshape(1,4))
        action = np.argmax(predicted[0])
  # Selecione uma ação aleatóriamente.
    else:
        action = np.random.randint(0, state_size)
    return action

def store_transition(c_state, action, reward, next_s):
        replay_memory.append((c_state, action, reward, next_s))
        
def targetnet_weight_update():
        target_network.set_weights(prediction_network.get_weights())
        
def training(batch_size):
    minibatch = random.sample(replay_memory, batch_size)

    for c_state, action, reward, next_s in minibatch: 
        target = reward 
        if not done:
                t = target_network.predict(next_s.reshape(1,4))[0]
                target = reward + discount_rate * np.amax(t) 
        
        target_f = prediction_network.predict(c_state.reshape(1,4))
        target_f[0][action] = target 
        prediction_network.fit(c_state.reshape(1,4), target_f, epochs=1, verbose=0)
    #print("Acao otima para cada canal:", target_f[0], "\n")
    return prediction_network
```
